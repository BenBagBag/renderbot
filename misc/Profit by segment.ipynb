```python
import pandas as pd

import bokeh.charts
import bokeh.charts.utils
import bokeh.io
import bokeh.models
import bokeh.palettes
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
df_store = pd.read_csv('store_data.csv', encoding = "ISO-8859-1", parse_dates=['Order Date'])
```

```python
df_store.head()
```

```python
# Add columns for sales where profit is plus or zero to negative
df_sales = df_store
df_sales['Profitable'] = df_store['Sales'].where(df_store['Profit'] > 0, 0)
df_sales['Unprofitable'] = df_store['Sales'].where(df_store['Profit'] <= 0, 0)
df_sales.head()
```

```python
df_sales = df_sales[['Order Date', 'Customer Segment', 'Profit', 'Sales', 'Profitable', 'Unprofitable']]
df_sales.groupby(['Customer Segment', 'Order Date']).sum()
df_sales.set_index('Order Date').groupby('Customer Segment').resample('M').sum()
```

```python
df_date_profit_by_segment = pd.DataFrame(df_store.groupby(['Customer Segment', 'Order Date'])['Profit'].sum())
df_date_profit_by_segment = df_date_profit_by_segment.reset_index(drop=False)
df_date_profit_by_segment.head()
```

```python
df_monthly_segment_profit = df_date_profit_by_segment.set_index('Order Date').groupby('Customer Segment').resample('M').sum()
df_monthly_segment_profit = df_monthly_segment_profit.reset_index(drop=False)
df_monthly_segment_profit.head()
```

```python
df_seg_prof = df_monthly_segment_profit[['Customer Segment', 'Profit']]
df_seg_prof.head()
```

```python
consumer_dict = {}
for index, row in df_seg_prof.iterrows():
    if row['Customer Segment'] in consumer_dict:
        consumer_dict[row['Customer Segment']].append(row['Profit'])
    else:
        consumer_dict[row['Customer Segment']] = [row['Profit']]
```

```python
consumer_dict
```

```python
from bokeh.charts import Area, show
area = Area(consumer_dict, title="Profit by Segment", legend="top_left",
            xlabel='Time', ylabel='Profit', plot_width=700, plot_height=400)

show(area)
```

```python
from bokeh.charts import Area, show
area_stacked = Area(consumer_dict, title="Stacked Area - Profit by Segment", legend="top_left",
            xlabel='Time', ylabel='Profit', plot_width=700, plot_height=400,
                   stack=True)

show(area_stacked)
```
